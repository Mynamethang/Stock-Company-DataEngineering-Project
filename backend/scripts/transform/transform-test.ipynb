```python
import pandas as pd
import os
import datetime
```

```python
def get_latest_json_file(directory):
    # Get a list of all JSON files in the directory
    json_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.json') and os.path.isfile(os.path.join(directory, f))]
    
    # Check if there are any JSON files in the directory
    if not json_files:
        return None, None
    
    # Get the latest JSON file based on modification time
    latest_file = max(json_files, key=os.path.getmtime)
    
    return latest_file
```
